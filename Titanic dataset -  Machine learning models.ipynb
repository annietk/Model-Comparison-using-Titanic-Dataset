```python
titanic['Embarked'].fillna(str(titanic['Embarked'].mode().values[0]),inplace=True)
pd.isnull(titanic).sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic.Sex = le.fit_transform(titanic.Sex)
titanic.Embarked = le.fit_transform(titanic.Embarked)
```

```python
X = titanic[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42, stratify = y)

X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
# Use a function for model-evaluation 
def model_evaluation(model):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)

    if model == Lin_Reg:
        model_name = "Linear Regression"
        y_pred = np.round(abs(y_pred),0)
    elif model == Log_Reg:
        model_name = "Logistic Regression"
    elif model == Deci_Tree:
        model_name = "Decision Tree Classification"
    else:
        model_name = "Random Forest Classification"
    
    print(y_pred)
     
    acc = accuracy_score(y_test, y_pred)
       
    cm = confusion_matrix(y_test, y_pred)
        
    cl_report = classification_report(y_test,y_pred)
    print("Confusion Matrix for "  + model_name + " :")  
    print(cm)
    print("Classification Report "  + model_name + " :") 
    print(cl_report)
    print("Accuracy Score for " + model_name + " :")   
    print(acc)
    return acc

models_dict = {}    # defined a dictionary for all storing the scores of all models 
```

```python
# 1. Liner Regression model

from sklearn.linear_model import LinearRegression
Lin_Reg = LinearRegression()

LinReg = model_evaluation(Lin_Reg)

models_dict['LinReg'] = {'model':"Linear Regression",
                         'score':LinReg}
```

```python
#2. Logistic Regression

from sklearn.linear_model import LogisticRegression
Log_Reg = LogisticRegression()

LogReg = model_evaluation(Log_Reg)
models_dict['LogReg'] = {'model':"Logistic Regression",
                         'score':LogReg}
```

```python
#3. DecisionTree Classifier
from sklearn.tree import DecisionTreeClassifier
Deci_Tree = DecisionTreeClassifier(criterion='entropy', random_state=0)

DeciTree = model_evaluation(Deci_Tree)
models_dict['DeciTree'] = {'model':"Decision Tree Classification",
                         'score':DeciTree}
```

```python
#4 RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
Ran_Forest = RandomForestClassifier(n_estimators = 10,criterion = 'entropy', random_state= 0)

RandFor = model_evaluation(Ran_Forest)
models_dict['RandFor'] = {'model':"Random Forest Classification",
                          'score':RandFor}
```
